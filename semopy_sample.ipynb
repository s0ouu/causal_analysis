```python
import semopy
from semopy import Model
```

```python
data = semopy.examples.political_democracy.get_data()
# import pandas as pd
# data = pd.read_csv('./data/sample.csv')
print(data.head())
```

```python
desc = semopy.examples.political_democracy.get_model()
# desc = open('model_desc.txt', 'r')
print(desc)
```

```python
model = Model(desc)
result = model.fit(data)
print(result)
```

```python
inspect = model.inspect(std_est=True)
print(inspect)
```

```python
stats = semopy.calc_stats(model)
print(stats.T)
```
